## Exame Nacional do Ensino Médio
The <b>Exame Nacional do Ensino Médio</b> is a standardized Brazilian national exam for high-schoolers. It is the biggest such exam in Brazil and the second largest in the world. It is similar to the French Baccalauréat or the US SAT. It is one of the main ways to gain admission into universities in Brazil. The test consits of 180 multiple choice questions and an essay. The questions are divided in four areas: Mathematics, Languages, Codes and related technologies, Natural Sciences (Physics, Chemistry and Biology) and Human Sciences (Geography, History, Philosophy and Sociology). We will use the data for the 2019 exam.

## Data source

https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem

<h2 style="color:green">6. The meritocracy problem</h2>

***

Let us analyze the distribution plots of scores separated by social class.

```python
ax=sns.displot(data=dados, x='NU_NOTA_CH',hue='FAIXA_RENDA',kind='kde',fill=True,palette=sns.color_palette('bright')[:5],hue_order=['E','D','C','B','A'])
ax.set_xlabels("Human Sciences Scores")
```

We can see in the distribution plot above how the score distribution increases for higher social classes. In order to have a better grasp on each distribution, let's plot for classes E and A separately.

```python
ax = sns.displot(data=dados[dados.FAIXA_RENDA=='A'], x='NU_NOTA_CH',kind='kde')
ax.set_xlabels("Human Sciences score (A)");
```

```python
ax = sns.displot(data=dados[dados.FAIXA_RENDA=='E'], x='NU_NOTA_CH',kind='kde')
ax.set_xlabels("Human Sciences score (E)");
```

***
Note that not only the median is higher for class A, but also the distribution is "thinner". In technical terms, the kurtosis is higher. Not only will applicants from class A have a higher median score, but also the variance is lower, meaning the chances of an applicant from class A having a lower score than the median is less than that of class E.

We can create a dataframe for all of the kurtosis calculations:
***

```python
faixas = ['E','D','C','B','A']
kurtosis={
    'NU_NOTA_CH':[],
    'NU_NOTA_CN':[],
    'NU_NOTA_MT':[],
    'NU_NOTA_LC':[],
    'NU_NOTA_REDACAO':[]
}

for key in kurtosis:
    for i in faixas:
        kurtosis[key].append(dados[dados.FAIXA_RENDA==i][key].kurtosis().round(3))
```

```python
kurtosis_df = pd.DataFrame(kurtosis,index=faixas)
kurtosis_df.columns=['Human Sciences','Natural Sciences','Mathematics','Languages and Codes','Essay']
```

```python
plt.figure(figsize=(14,12))
g=sns.heatmap(kurtosis_df.T, annot=True, fmt=".3f",cmap='coolwarm')
g.set_yticklabels(labels=['Human Sciences','Natural Sciences','Mathematics','Languages and Codes','Essay'],rotation=0)
g.set_title("Score kurtosis increase per social class, except for mathematics", fontsize = 24, loc='left')
g=g
```

***
The image below shows a galton board, a device used to show that given a big enough number of samples, the binomial distribution approximates to a normal distribution. Under a meritocratic view, each time a ball bumps into a peg, it would simbolize the effort or talent of that individual, meaning that in order to reach the right-most bins, it would require personal effort to do so.

However, as we saw in the distribution plots above, in Brazil's reality the distribution is shifted depending on social class. This would be similar as to having the canal on top be shifted, or the whole board being inclined to the side. Furthermore, the change in kurtosis would mean that the pegs are configured in such a way that the balls have less dispersion depending on the class.

As Leonard Mlodinow describes in his book The Drunkard's Walk, in the real world, the events that these pegs represent are, more often than we'd like to believe, being controled by randomness.

Adding to the above, we have also seen throughout this notebook the various aspects of inequality, not only for social class.

Therefore, can we really say that the education and access to universities in Brazil is meritocratic?
